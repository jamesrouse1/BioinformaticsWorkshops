## Part 3: Count Matrix

```python
df_counts = pd.DataFrame({
    'Gene': ['MYC', 'KRAS', 'TP53', 'EGFR', 'CDKN2A', 'SMAD4', 'MKI67', 'GATA6'],
    'Tumor_1': [500, 900, 400, 700, 100, 220, 650, 150],
    'Tumor_2': [520, 880, 390, 680, 120, 210, 640, 160],
    'Tumor_3': [510, 910, 410, 720, 110, 205, 670, 155],
    'Tumor_4': [530, 905, 405, 710, 115, 215, 660, 145],
    'Normal_1': [200, 300, 450, 250, 260, 500, 140, 420],
    'Normal_2': [210, 320, 470, 240, 270, 510, 150, 410],
    'Normal_3': [205, 310, 460, 260, 265, 495, 145, 430],
    'Normal_4': [215, 305, 455, 255, 275, 505, 155, 415]
})
df_counts
```

```python
print(df_counts.shape)
df_counts_indexed = df_counts.set_index('Gene')
df_counts_indexed
```

## Part 4: Indexing Practice

```python
print(df_counts_indexed.iloc[:, 0])
print(df_counts_indexed.iloc[0, :])
print(df_counts_indexed.iloc[:, 0:4])
print(df_counts_indexed.iloc[:, 4:8])
print(df_counts_indexed.loc['KRAS', 'Tumor_3'])
print(df_counts_indexed.loc['GATA6', 'Normal_2'])
df_counts_indexed.info()
df_counts_indexed.describe()
```

## Part 5: CPM Normalization

```python
total_counts = df_counts_indexed.sum(axis=0)
print(total_counts)

df_cpm = df_counts_indexed.div(total_counts, axis=1) * 1e6
df_cpm
```

```python
df_cpm.loc['MYC','Tumor_1']
```

```python
df_cpm.loc['GATA6','Normal_4']
```

## Part 6: Tumor vs Normal Means

```python
df_cpm['Normal_Mean'] = df_cpm.iloc[:, 4:8].mean(axis=1)
df_cpm['Tumor_Mean'] = df_cpm.iloc[:, 0:4].mean(axis=1)
df_cpm['Difference'] = df_cpm['Tumor_Mean'] - df_cpm['Normal_Mean']
df_cpm['Rank'] = df_cpm['Difference'].rank(ascending=False)
df_cpm.sort_values('Difference', ascending=False)
```

## Part 7: Fold Change

```python
df_cpm['Fold_Change'] = (df_cpm['Tumor_Mean'] + 1) / (df_cpm['Normal_Mean'] + 1)
df_cpm['Log2_FC'] = np.log2(df_cpm['Fold_Change'])
df_cpm[['Tumor_Mean', 'Normal_Mean', 'Fold_Change', 'Log2_FC']].sort_values('Log2_FC', ascending=False)
```

## Part 8: T-Test Practice

```python
tumor_vals = df_cpm.iloc[:, 0:4]
normal_vals = df_cpm.iloc[:, 4:8]

ttest_results = ttest_ind(tumor_vals.values, normal_vals.values, axis=1, equal_var=False)
df_cpm['p_value'] = ttest_results.pvalue

df_cpm['Label'] = 'Not Significant'
df_cpm.loc[(df_cpm['p_value'] < 0.05) & (df_cpm['Log2_FC'] > 0), 'Label'] = 'Up in Tumor'
df_cpm.loc[(df_cpm['p_value'] < 0.05) & (df_cpm['Log2_FC'] < 0), 'Label'] = 'Up in Normal'

df_cpm.sort_values('p_value')
```

## Part 9: New Practice That Was Not in the Tutorial

Tasks:
1. Add a column called `Tumor_SD` using the 4 tumor CPM columns.
2. Add a column called `Normal_SD` using the 4 normal CPM columns.
3. Add a column called `Range_CPM` equal to the row max minus the row min across all 8 CPM sample columns.
4. Sort the table by `Tumor_SD` from high to low.
5. Which gene has the largest `Tumor_SD`?
6. Which gene has the largest `Range_CPM`?

Guidance:
- use `.std(axis=1)` for row-wise standard deviation
- use `.max(axis=1)` and `.min(axis=1)` for row-wise range
- do these calculations on the CPM matrix, not the raw counts
- for tumor-only calculations, use the 4 tumor CPM columns
- for normal-only calculations, use the 4 normal CPM columns


```python
df_cpm['Tumor_SD'] = df_cpm.iloc[:, 0:4].std(axis=1)
df_cpm['Normal_SD'] = df_cpm.iloc[:, 4:8].std(axis=1)
df_cpm['Range_CPM'] = df_cpm.iloc[:, 0:8].max(axis=1) - df_cpm.iloc[:, 0:8].min(axis=1)

df_cpm[['Tumor_SD', 'Normal_SD', 'Range_CPM']].sort_values('Tumor_SD', ascending=False)
```
